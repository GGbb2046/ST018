```python
import pandas as pd #importing panda module
import matplotlib.pyplot as plt #importing matplotlib for graphical plot
fb = pd.read_csv("FB.csv")# reading the downloaded yahoo csv finance file  
cs = pd.read_csv("CSCO.csv")
ib = pd.read_csv("IBM.csv")
go = pd.read_csv("goog.csv")
d = fb.Date # calling date as a 
f = (fb.Close - fb.Close.shift(1)).round(2)# creating a data series of the difference of the closing stock prices of the respective company
c = (cs.Close - cs.Close.shift(1)).round(2)
i = (ib.Close - ib.Close.shift(1)).round(2)
g = (go.Close - go.Close.shift(1)).round(2)
```

```python
# Defining colors for reflecting loss in red and gain in green
def color(col):
    if col < 0:
        color = 'red'
    elif col > 0:
        color = 'green'
    else:
        color = 'black'

    return 'color: %s' % color
```

```python
df = pd.DataFrame([d,f,c,i,g]).T #creating a dataframe from the above data series
df = df.iloc[1:] #removing the first row with null data
df.columns = ['Date','Facebook', 'Cisco', 'IBM', 'Google'] #assigning new column names
df['Total Daily Gain/(Loss)'] = (df['Facebook']+df['Cisco']+df['IBM']+df['Google']) # creating a new column for total gain/loss
dfx = df.set_index('Date') # setting the date as the index column 
dfy = dfx.style.applymap(color, subset=['Total Daily Gain/(Loss)'])#setting the color function for total gain/loss column.
dfy #output
```

```python
%matplotlib inline
dfx.plot.line(figsize=(15,10))
```
